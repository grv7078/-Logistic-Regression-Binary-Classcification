```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset.drop(columns=["User ID","Gender","EstimatedSalary"],inplace=True)
dataset.head(5)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age",y="Purchased",data=dataset)
plt.show()
```

```python
x=dataset[["Age"]]
y=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[40]])
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age",y="Purchased",data=dataset)
sns.lineplot(x="Age",y=lr .predict(x),data=dataset,c="red")
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age",y="Purchased",data=dataset)
sns.scatterplot(x="Age",y=lr .predict(x),data=dataset,c="red")
plt.show()
```
